01. <a name="id_1">Quais os 10 livros com maiores avaliações?</a>

```python
ps.sqldf("""
SELECT
  titulo,
  autor,
  avaliacao
FROM
  livros
ORDER BY
  avaliacao DESC
LIMIT
  10
""")
```

02. <a name="id_2">Quais editoras possuem mais livros publicados?</a>

```python
ps.sqldf("""
SELECT
  editora,
  COUNT(titulo) AS qtd_livros
FROM
  livros
GROUP BY
  editora
ORDER BY
  qtd_livros DESC
LIMIT
  10
""")
```

03. <a name="id_3">Quantidade de livros publicados por ano por editora</a>

```python
ps.sqldf("""
SELECT
  ano,
  editora,
  COUNT(titulo) AS qtd_livros
FROM
  livros
GROUP BY
  editora,
  ano
ORDER BY
  ano DESC,
  qtd_livros DESC
LIMIT
  10
""")
```

```python
# Curiosidade: Buscando livros da editora "The"

ps.sqldf("""
SELECT
  *
FROM
  livros
WHERE
  editora = 'The'
""")
```

04. <a name="id_4">Qual livro foi escrito em 1193</a>

Constatou-se que o livro lançado em 1193 aparenta ser um erro no dataset

```python
ps.sqldf("""
SELECT
  *
FROM
  livros
WHERE
  ano = 1193
""")
```

05. <a name="id_5">Qual autor mais lido por gênero (Male / Female)?</a>

```python
#Female

ps.sqldf("""
SELECT
  autor,
  COUNT(titulo) as qtd_livros
FROM
  livros
WHERE
  female = 100
GROUP by
  autor
ORDER BY
  qtd_livros DESC
LIMIT
  20
""")
```

```python
#Male

ps.sqldf("""
SELECT
  autor,
  COUNT(titulo) as qtd_livros
FROM
  livros
WHERE
  male = 100
GROUP by
  autor
ORDER BY
  qtd_livros DESC
LIMIT
  20
""")
```

06. <a name="id_6">Quais os 10 livros com maior número de abandono?</a>

```python
ps.sqldf("""
SELECT DISTINCT
  titulo,
  abandonos
FROM
  livros
ORDER BY
  abandonos DESC
LIMIT
  10
""")
```

07. <a name="id_7">Cinco editoras com maior média de abandono</a>

```python
ps.sqldf("""
SELECT
  editora,
  AVG(abandonos) AS media_abandono
FROM
  livros
GROUP BY
  editora
ORDER BY
  media_abandono DESC
LIMIT
  5
""")
```

08. <a name="id_8">Quantidade de páginas escritas por autor</a>

```python
ps.sqldf("""
SELECT
  autor,
  SUM(paginas) as total_paginas
FROM
  livros
GROUP BY
  autor
ORDER BY
  total_paginas DESC
LIMIT
  10
""")
```

09. <a name="id_9">Ordem dos livros melhores avaliados por quantidade avaliação</a>

```python
ps.sqldf("""
SELECT
  titulo,
  autor,
  rating,
  avaliacao
FROM
  livros
ORDER BY
  avaliacao DESC,
  rating DESC
LIMIT
  25
""")
```

10. <a name="id_10">Melhores livros por gênero literário</a>

```python
#Ficção
ps.sqldf("""
SELECT
  titulo,
  autor,
  genero,
  rating
FROM
  livros
WHERE
  genero LIKE '%%Ficção%%'
ORDER BY
  rating DESC
LIMIT
  25
""")
```
