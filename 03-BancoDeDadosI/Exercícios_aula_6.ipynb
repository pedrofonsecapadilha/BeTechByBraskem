### Exercício 6:

### Exercício 8:

Query que retorne os 10 primeiros alunos que ingressaram em seus cursos;

```python
ps.sqldf("""
SELECT
  primeiro_nome,
  ultimo_nome,
  data_ingresso
FROM
  alunos
ORDER BY
  data_ingresso ASC
LIMIT
  10
""")
```

### Exercício 9:

### Exercício 10:

Query que retorne apenas os alunos que moram em casas de 3 dígitos;

```python
ps.sqldf("""
SELECT
  cpf,
  primeiro_nome,
  numero,
  CAST(numero AS text) AS numero_texto,
  LENGTH(CAST(numero AS text)) AS digitos
FROM
  alunos
WHERE
  LENGTH(CAST(numero AS text)) = 3
""")
```

### Exercício 13:

Query que retorne apenas alunos que moram em cidades com 7 ou mais letras;

```python
ps.sqldf("""
SELECT
  A.primeiro_nome,
  A.ultimo_nome,
  C.cidade,
  C.estado
FROM
  alunos AS A
  LEFT JOIN cidades AS C
  ON A.id_cidade = C.id_cidade
WHERE
  LENGTH(CAST(cidade AS text)) >= 7
""")
```

### Exercício 14:

Query que retorne apenas alunos com primeira letra do nome e do sobrenome sendo vogais;

```python
ps.sqldf("""
SELECT
  primeiro_nome,
  ultimo_nome
FROM
  alunos
WHERE
  (
  primeiro_nome LIKE "a%" OR
  primeiro_nome LIKE "e%" OR
  primeiro_nome LIKE "i%" OR
  primeiro_nome LIKE "o%" OR
  primeiro_nome LIKE "u%"
  )
  AND
  (
  ultimo_nome LIKE "a%" OR
  ultimo_nome LIKE "e%" OR
  ultimo_nome LIKE "i%" OR
  ultimo_nome LIKE "o%" OR
  ultimo_nome LIKE "u%"
  )
""")
```

### Exercício 15:

Query que retorne apenas Baianos com nomes com mais de 4 letras;

```python
ps.sqldf("""
SELECT
  A.primeiro_nome,
  A.ultimo_nome,
  C.cidade,
  C.estado
FROM
  alunos AS A
  LEFT JOIN cidades AS C
  ON A.id_cidade = C.id_cidade
WHERE
  estado LIKE "BA"
""")
```
