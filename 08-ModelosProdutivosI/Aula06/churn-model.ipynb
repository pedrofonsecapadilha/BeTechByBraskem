```python
# Tratar valores ausentes

df = df.dropna()
```

```python
# Transformar variáveis categóricas em variáveis dummy

df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
# Dividir o conjunto de dados de treinamento e teste

X = df.drop("Churn_Yes", axis=1)
y = df["Churn_Yes"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Normalizar os dados

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Inferir e avaliar o modelo

y_pred = model.predict(X_test)
print("Acurácia: ", accuracy_score(y_test, y_pred).round(2))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred))
```

```python
# Registrar o modelo e métricas dentro do MLflow

import os

os.environ["MLFLOW_TRACKING_URI"] = "http://127.0.0.1:5000"
os.environ["MLFLOW_ARTIFACTS_URI"] = "./artifacts"

with mlflow.start_run() as run:
    mlflow.log_param("modelo", "Regressão Logística")
    mlflow.log_metric("acuracia", accuracy_score(y_test, y_pred))
    mlflow.sklearn.log_model(model, "modelo_logistic_regression")

    run_id = run.info.run_id
    model_uri = f"runs:/{run_id}/modelo"
    mlflow.register_model(model_uri, "ModeloChurn")
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predito")
plt.ylabel("Real")
plt.title("Matriz de Confusão")
plt.show()
```
