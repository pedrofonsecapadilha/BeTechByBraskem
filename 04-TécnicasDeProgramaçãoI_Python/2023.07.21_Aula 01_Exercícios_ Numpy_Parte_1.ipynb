2. Crie um array 1-dimensional (com 100 elementos) com números reais (float) aleatórios, e retorne as seguintes métricas:
  - média
  - mediana
  - desvio padrão
  - máximo
  - mínimo
  - índice do elemento máximo
  - indice do elemento mínimo

```python
aleatorio2 = np.random.rand(100)
print(aleatorio2)
```

```python
print(f"Média: {aleatorio2.mean()}")
print(f"Mediana: {np.median(aleatorio2)}")
print(f"Desvio Padrão: {aleatorio2.std()}")
print(f"Máximo: {aleatorio2.max()}")
print(f"Mínimo: {aleatorio2.min()}")
print(f"Índice Min: {aleatorio2.argmin()}")
print(f"Índice Max: {aleatorio2.argmax()}")
```

3. Crie uma array 2-dimensional (10x5) com números reais (float) aleatórios, e retorne as seguintes métricas:
  - média
  - mediana
  - desvio padrão
  - máximo
  - mínimo
  - índice do elemento máximo
  - indice do elemento mínimo

```python
aleatorio3 = np.random.rand(50)
matriz = aleatorio3.reshape((10, 5))
print(matriz)
```

```python
print(f"Média: {matriz.mean()}")
print(f"Mediana: {np.median(matriz)}")
print(f"Desvio Padrão: {matriz.std()}")
print(f"Máximo: {matriz.max()}")
print(f"Mínimo: {matriz.min()}")
print(f"Índice Max: {matriz.argmax()}")
print(f"Índice Min: {matriz.argmin()}")
```

4. Crie um array 1-dimensional com 100 inteiros aleatórios de 0 a 1000, e substitua cada valor ímpar por -1.

```python
arr4 = np.random.randint(0, 1001, size=100)
print(arr4)
```

```python
arr4[arr4 % 2 != 0] = -1
print(arr4)
```
