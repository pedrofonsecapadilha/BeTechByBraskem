1. Leia a base de dados titanic usando numpy e realize as seguintes análises:

- Quantidade de idades (Age) com valor faltante.

- Quantidade de passageiro com zero SibSp.

- Calcule a média da idade dos passageiros.

- Verifique em quais classes (Pclass) os passageiros estavam.

- Calcule a quantidade de sobreviventes ao desastre (Survived).

- Das classes Pclass=3, quantos pessoas do sexo feminino estavam presentes?

```python
titanic = np.loadtxt(
    "datasets/train2.csv",
    delimiter=',',
    skiprows=1,
    dtype=[
      ('PassengerId', int),
      ('Survived', int),
      ('Pclass', int),
      ('Name', 'U150'),
      ('Sex', 'U6'),
      ('Age', 'U5'),
      ('SibSp', int),
      ('Parch', int),
      ('Ticket', 'U15'),
      ('Fare', float),
      ('Cabin', 'U5'),
      ('Embarked', 'U1')
    ],
)
print(titanic[:5])
```

```python
# Quantidade de idades (Age) com valor faltante.

soma = titanic['Age'] == ""
np.sum( soma == True)
```

```python
# Quantidade de passageiro com zero SibSp

sibsp_0 = titanic['SibSp'] == 0
np.sum( sibsp_0 == True)
```

```python
# Calcule a média da idade dos passageiros.

idades = []
for e in titanic['Age']:
  if e != '':
    idades.append(float(e))

print(f"Média das Idades: {np.array(idades).mean()}")
```

```python
# Verifique em quais classes (Pclass) os passageiros estavam.

np.unique(titanic['Pclass'])
```

```python
# Calcule a quantidade de sobreviventes ao desastre (Survived).

sobreviventes = titanic['Survived'] == 1
np.sum(sobreviventes == True)
```

```python
# Das classes Pclass=3, quantos pessoas do sexo feminino estavam presentes?

sibsp_3_F = titanic[(titanic["Pclass"] == 3) & (titanic["Sex"] == 'female')]
type(sibsp_3_F)
sibsp_3_F.shape[0]
# len(sibsp_3_F)
```
