## Dendrograma

Crie o dendrograma deste dataset usando ligação completa e responda no forms quantos grupos teremos ao usarmos como corte o valor de distância euclidiana (eixo y) como 3.

```python
X = df.iloc[:, [0,1,2, 3]].values
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(16,6))
dendrograma = sch.dendrogram(sch.linkage(X, method='complete'), leaf_font_size=8, color_threshold=3)

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distância Euclidiana')
plt.show()
```

## Cluster Aglomerativo

Treine o modelo de cluster aglomerativo com a quantidade de clusters encontrada no exercício anterior, método de cálculo de distância (affinity) como euclidiana e tipo de ligação como completa.

```python
from sklearn.cluster import AgglomerativeClustering

agg_cl = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
agg_cl.fit(X)
```
