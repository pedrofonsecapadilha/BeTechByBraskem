## DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.4, min_samples=3) # raio de 0.4 e mínimo de amostras de 3
dbscan.fit(df)
```

```python
dbscan.labels_
```

```python
df['clusters'] = dbscan.labels_
```
